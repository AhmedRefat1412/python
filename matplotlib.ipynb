Advanced plot
-------------

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100) # بيجيب 100 رقم بين  1 ل 10 بترتيب 
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/3) * np.sin(2 * x)

# Create figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# # Plot primary y-axis data
ax1.plot(x, y1, color='blue', label='sin(x)', linewidth=2)
ax1.plot(x, y2, color='green', linestyle='--', label='cos(x)')
ax1.set_xlabel('Time (s)', fontsize=12)
ax1.set_ylabel('Amplitude', fontsize=12)

# هنا هيغير لون احداثيات محور الصادات الي الاسود
ax1.tick_params(axis='y', labelcolor='black') 

# Add a second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y3, color='red', label='damped sin(2x)', linewidth=2)
ax2.set_ylabel('Damped Wave', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add grid, title # اضافات مربعات في خلفيه الشكل و عنوان الشكل العام
ax1.grid(True, linestyle=':', linewidth=0.5)
ax1.set_title('Advanced Plot with Twin Axes and Annotations', fontsize=14)

# Annotate a specific point اضافه تعليق
ax1.annotate('Peak here' # الاسم 
             , xy=(1.57, 1), # احداثيات النقطه الي هبقا عليها التعليق
               xytext=(3, 1.2),# احداثيات المكان الي عليه التكست
             arrowprops=dict(color='red', arrowstyle='->'), # تحديد نوع السهم ولونه 
             fontsize=10, backgroundcolor='white')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right')

plt.tight_layout()
plt.show()
```

انشاء 3 محور صادات

```python
import matplotlib.pyplot as plt

# البيانات
x = [1, 2, 3, 4]
y1 = [5, 10, 30, 40]  # القيم الكبيرة
y2 = [1, 2, 3, 4]      # القيم الصغيرة
y3 = [5, 10, 15, 60]   # القيم المتوسطة

# إنشاء الشكل وتحديد ججمه 4*4 مثلا  والمحاور
fig, ax1 = plt.subplots(figsize=(4, 4))

# رسم الخط الأول (الخط الأزرق)
ax1.plot(x, y1, color='blue', marker='o', label='Large Values (Blue)')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Big Values', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# إنشاء محور Y ثاني
ax2 = ax1.twinx()
# رسم الخط الثاني (الخط الأحمر)
ax2.plot(x, y2, color='red', marker='x', label='Small Values (Red)')
ax2.set_ylabel('Small Values', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# إنشاء محور Y ثالث
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # تحريك المحور الثالث بعيدًا عن الثاني
# رسم الخط الثالث (الخط الأخضر)
ax3.plot(x, y3, color='green', marker='s', label='Medium Values (Green)')
ax3.set_ylabel('Medium Values', color='green')
ax3.tick_params(axis='y', labelcolor='green')

# إضافة عنوان
plt.title('Three Y-Axis Plot')

# عرض الرسم البياني
plt.show()
```
