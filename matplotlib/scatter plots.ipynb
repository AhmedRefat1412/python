scatter plots 
-----------

```python
import matplotlib .pyplot as plt

# الداتا
x=[2,3,6,4,8,5,4,6,2]
y=[2,6,8,9,7,2,5,6,1]

plt.figure(figsize=(6,5))
plt.scatter(x,y ,color='red',marker='o',s=20)

plt.title('scatter plots')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

advanced example
-----------------
by load dataset from seaborn (tips dataset)

```python
import seaborn as sns
import matplotlib.pylab as plt
df = sns.load_dataset('tips')  # داتا حقيقية من مكتبة seaborn

plt.figure(figsize=(8, 5))
plt.scatter(df['total_bill'], df['tip'], 
            alpha=0.7,  #هي درجه تباين النقاط وبتيقا من 0ل 1  alpha 
            color='green',# لون النقطه
            edgecolors='black', # الحواف حوالين النقطه 
             )
plt.title('Tips vs Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid()
plt.show()
```
