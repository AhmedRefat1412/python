```python
import matplotlib.pyplot as plt
import numpy as np

# Data
subjects = ['Math', 'Science', 'History', 'Art']
term1_scores = [85, 90, 75, 80]
term2_scores = [88, 85, 78, 82]

# Bar positions
x = np.arange(len(subjects))  # [0, 1, 2, 3]
bar_width = 0.3 # ÿπÿ±ÿ∂ ÿßŸÑÿ®ÿßÿ± 

# Create plot
fig, ax = plt.subplots(figsize=(8, 5)) # ÿ¥ŸÉŸÑ ÿßŸÑÿ®ÿßÿ± Ÿàÿ≠ÿ¨ŸÖŸá

# Plot the bars
bars1 = ax.bar(x-bar_width/2, term1_scores, width=bar_width, label='Term 1', color='skyblue')
bars2 = ax.bar(x + bar_width/2, term2_scores, bar_width, label='Term 2', color='salmon')

# Add labels and title
ax.set_xlabel('Subjects')
ax.set_ylabel('Scores')
ax.set_title('Student Scores: Term 1 vs Term 2')
ax.set_xticks(x) # ÿπÿ¥ÿßŸÜ ŸÉÿ™ÿ® ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸä ÿ™ÿ≠ÿ™ ŸÅŸä ŸÖÿ≠Ÿàÿ± ÿßŸÑÿ≥ŸäŸÜÿßÿ™ ŸáŸÜÿß ÿ®ÿπÿ±ŸÅŸá ÿπÿØÿØŸáŸÖ 
ax.set_xticklabels(subjects) # xticks ŸàŸáŸÜÿß ÿ®ŸÇŸàŸÑŸá ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸä Ÿáÿ™ÿ®ŸÇÿß ŸÖŸàÿ¨ŸàÿØŸá ŸÅŸä ŸÖÿ≠Ÿàÿ± ÿßÿ≥ŸäŸÜÿßÿ™ ÿßŸáŸä ŸàŸÑÿßÿ≤ŸÖ ÿ™ÿ®ŸÇÿß ŸÜŸÅÿ≥ ÿßŸÑÿπÿØÿØ ŸÖÿπ 
ax.legend()

# Show the plot
plt.tight_layout() # ÿπÿ¥ÿßŸÜ ŸäÿÆŸÑŸä ŸÉŸÑ ÿ≠ÿßÿ¨Ÿá Ÿàÿßÿ∂ÿ≠Ÿá ŸÅŸä ÿßŸÑÿ¥ŸÉŸÑ 
plt.show()
```

üî∂ Stacked Bar Chart ‚Äì ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸä ÿ®ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑŸÖÿ™ÿ±ÿßŸÉÿ®ÿ©
-----------
ÿ®ÿ≠ÿ∑ ÿπŸÖŸàÿØ ŸÅŸä ÿπŸÖŸàÿØ ÿ™ÿßŸÜŸä 

```python
import matplotlib.pyplot as plt

# ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
subjects = ['Math', 'Science', 'History', 'Art']
test_scores = [50, 55, 40, 45]
project_scores = [35, 30, 35, 37]

# ÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑŸÖÿ™ÿ±ÿßŸÉÿ®ÿ©
fig, ax = plt.subplots(figsize=(8, 5))

# ÿßŸÑÿπŸÖŸàÿØ ÿßŸÑÿ£ŸàŸÑ (ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±ÿßÿ™)
ax.bar(subjects, test_scores, label='Tests', color='skyblue')

# ÿßŸÑÿπŸÖŸàÿØ ÿßŸÑÿ´ÿßŸÜŸä (ÿßŸÑŸÖÿ¥ÿ±Ÿàÿπÿßÿ™) ‚Äî ŸÖÿ™ÿ±ŸÉÿ® ŸÅŸàŸÇ ÿßŸÑÿ£ŸàŸÑ
ax.bar(subjects, project_scores, bottom=test_scores, label='Projects', color='orange')

# ÿπŸÜŸàÿßŸÜ ŸàŸÖÿ≠ÿßŸàÿ±
ax.set_xlabel('Subjects')
ax.set_ylabel('Total Score')
ax.set_title('Stacked Bar Chart: Tests and Projects')
ax.legend()

plt.tight_layout()
plt.show()
```

ÿßŸÑÿ±ÿ≥ŸÖ ÿ®ÿ∑ÿ±ŸäŸÇŸá ÿßŸÅŸÇŸäŸá

plt .barh ÿ®ÿ™ÿÆŸÑŸä ÿßŸÑÿ¥ŸÉŸÑ ÿßŸÅŸÇŸä 

```python
import matplotlib.pyplot as plt

# ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
subjects = ['Math', 'Science', 'History', 'Art']
scores = [85, 90, 75, 80]

# ÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑÿ£ŸÅŸÇŸäÿ©
plt.barh(subjects, scores, color='lightgreen')

# ÿßŸÑÿπŸÜÿßŸàŸäŸÜ ŸàÿßŸÑŸÖÿ≠ÿßŸàÿ±
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Horizontal Bar Chart: Student Scores')

plt.tight_layout()
plt.show()
```
