🔷 Grouped Bar Chart in Matplotlib
----------------------------------------

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
subjects = ['Math', 'Science', 'History', 'Art']
term1_scores = [85, 90, 75, 80]
term2_scores = [88, 85, 78, 82]

# Bar positions
x = np.arange(len(subjects))  # [0, 1, 2, 3]
bar_width = 0.3 # عرض البار 

# Create plot
fig, ax = plt.subplots(figsize=(8, 5)) # شكل البار وحجمه

# Plot the bars
bars1 = ax.bar(x-bar_width/2, term1_scores, width=bar_width, label='Term 1', color='skyblue')
bars2 = ax.bar(x + bar_width/2, term2_scores, bar_width, label='Term 2', color='salmon')

# Add labels and title
ax.set_xlabel('Subjects')
ax.set_ylabel('Scores')
ax.set_title('Student Scores: Term 1 vs Term 2')
ax.set_xticks(x) # عشان كتب البيانات الي تحت في محور السينات هنا بعرفه عددهم 
ax.set_xticklabels(subjects) # xticks وهنا بقوله القيم الي هتبقا موجوده في محور اسينات اهي ولازم تبقا نفس العدد مع 
ax.legend()

# Show the plot
plt.tight_layout() # عشان يخلي كل حاجه واضحه في الشكل 
plt.show()
```

🔶 Stacked Bar Chart – الرسم البياني بالأعمدة المتراكبة
-----------
بحط عمود في عمود تاني 

```python
import matplotlib.pyplot as plt

# البيانات
subjects = ['Math', 'Science', 'History', 'Art']
test_scores = [50, 55, 40, 45]
project_scores = [35, 30, 35, 37]

# رسم الأعمدة المتراكبة
fig, ax = plt.subplots(figsize=(8, 5))

# العمود الأول (الاختبارات)
ax.bar(subjects, test_scores, label='Tests', color='skyblue')

# العمود الثاني (المشروعات) — متركب فوق الأول
ax.bar(subjects, project_scores, bottom=test_scores, label='Projects', color='orange')

# عنوان ومحاور
ax.set_xlabel('Subjects')
ax.set_ylabel('Total Score')
ax.set_title('Stacked Bar Chart: Tests and Projects')
ax.legend()

plt.tight_layout()
plt.show()
```

الرسم بطريقه افقيه 
-----------
plt .barh بتخلي الشكل افقي 

```python
import matplotlib.pyplot as plt

# البيانات
subjects = ['Math', 'Science', 'History', 'Art']
scores = [85, 90, 75, 80]

# رسم الأعمدة الأفقية
plt.barh(subjects, scores, color='lightgreen')

# العناوين والمحاور
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Horizontal Bar Chart: Student Scores')

plt.tight_layout()
plt.show()
```
