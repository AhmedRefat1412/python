pie chart
-----------

```python
import matplotlib.pyplot as plt

# البيانات
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 25, 25, 20]  # النسب المئوية أو القيم
colors = ['red', 'yellow', 'pink', 'brown']
explode = [0.1, 0, 0, 0]  # لتكبير شريحة معينة (اختياري)

# إنشاء الرسم
plt.pie(sizes, labels=labels,
         colors=colors, # الون الاشكال الي انا محددها فوق
         explode=explode, # مقدار الخروج عن مركز الدايره
         autopct='%1.1f%%', #  % اظهار علامه بلميه 
         shadow=True, # ظل حوالين الدايره
         startangle=140 # زاوية بداية أول شريحة
         )

plt.title('Fruit Distribution')
plt.axis('equal')  # تخلي الدائرة فعلاً دائرية

plt.legend( )

plt.show()
```

علي داتا ست سفينه تايتنك فمن خلال عدد الاسخاص الي في كل درجه pie chart انشاء 
-------------------------------------------

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# تحميل بيانات Titanic
df = sns.load_dataset('titanic')

# حساب عدد كل فئة في عمود 'class'
class_counts = df['class'].value_counts()
print(class_counts.index)

# إعداد الألوان
colors = ['#66b3ff', '#99ff99', '#ffcc99']

# رسم المخطط الدائري
plt.figure(figsize=(8, 8))
plt.pie(class_counts,
        labels=class_counts.index,
        colors=colors,
        autopct=lambda pct: f"{pct:.1f}%\n({int(pct*sum(class_counts)/100)})",
        startangle=90,
        shadow=True,
        explode=[0.05]*len(class_counts))

# تحسين شكل الرسم
plt.title('Passenger Distribution by Class - Titanic Dataset', fontsize=14)
plt.axis('equal')
plt.legend(title="Class", loc="upper left", bbox_to_anchor=(0.9, 1) )# بتحدد موقعها فين 

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# إنشاء بيانات عشوائية: أعمار 1000 شخص بين 10 و80 سنة
ages = np.random.randint(10, 80, 1000)

# رسم الهستوجرام
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=15, color='skyblue', edgecolor='red')

# تخصيصات الشكل
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.grid(True)
plt.show()
```
