```python
import matplotlib.pyplot as plt

# 
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 25, 25, 20]  #    
colors = ['red', 'yellow', 'pink', 'brown']
explode = [0.1, 0, 0, 0]  #    ()

#  
plt.pie(sizes, labels=labels,
         colors=colors, #      
         explode=explode, #     
         autopct='%1.1f%%', #  %    
         shadow=True, #   
         startangle=140 #    
         )

plt.title('Fruit Distribution')
plt.axis('equal')  #    

plt.legend( )

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#   Titanic
df = sns.load_dataset('titanic')

#       'class'
class_counts = df['class'].value_counts()
print(class_counts.index)

#  
colors = ['#66b3ff', '#99ff99', '#ffcc99']

#   
plt.figure(figsize=(8, 8))
plt.pie(class_counts,
        labels=class_counts.index,
        colors=colors,
        autopct=lambda pct: f"{pct:.1f}%\n({int(pct*sum(class_counts)/100)})",
        startangle=90,
        shadow=True,
        explode=[0.05]*len(class_counts))

#   
plt.title('Passenger Distribution by Class - Titanic Dataset', fontsize=14)
plt.axis('equal')
plt.legend(title="Class", loc="upper left", bbox_to_anchor=(0.9, 1) )#    

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#   :  1000   10 80 
ages = np.random.randint(10, 80, 1000)

#  
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=15, color='skyblue', edgecolor='red')

#  
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.grid(True)
plt.show()
```
